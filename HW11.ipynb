Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.loc[ratings.rating <= 2.0, 'Class'] = 'low'
ratings.loc[ratings.rating <= 4.0, 'Class'] = 'medium'
ratings.loc[ratings.rating >= 4.5, 'Class'] = 'high'
ratings.head()
```

Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность 
определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.


```python
keywords = pd.read_csv('keywords.csv')
keywords.tail()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
pd.DataFrame(geo_data)
```

```python
def geo(row):
    for row in row:
        if 'вк' in row.split():
            return 'центр'
```

```python
keywords['region'] = keywords['keyword'].apply(geo)
keywords.head(20)
```

```python
keywords.loc[keywords.keyword.str.contains('москва'), 'region'] = 'Центр'
keywords.loc[keywords.keyword.str.contains('тула'), 'region'] = 'Центр'
keywords.loc[keywords.keyword.str.contains('ярославль'), 'region'] = 'Центр'
keywords.loc[keywords.keyword.str.contains('петербург'), 'region'] = 'Северо-Запад'
keywords.loc[keywords.keyword.str.contains('псков'), 'region'] = 'Северо-Запад'
keywords.loc[keywords.keyword.str.contains('мурманск'), 'region'] = 'Северо-Запад'
keywords.loc[keywords.keyword.str.contains('владивосток'), 'region'] = 'Дальний Восток'
keywords.loc[keywords.keyword.str.contains('сахалин'), 'region'] = 'Дальний Восток'
keywords.loc[keywords.keyword.str.contains('хабаровск'), 'region'] = 'Дальний Восток'
keywords.region.count()
```
